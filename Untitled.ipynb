## Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Import DataSet

**Importing Test Data**

```python
df = pd.read_csv("test_3cot.csv")
df
```

```python
df.info()
```

```python
df.describe()
```

**Converting Object Type to DateTime using to_datatime function**

```python
df["t"] = pd.to_datetime(df["t"])
df.info()
```

**Removing Index column and Set Index to Time/Date column**

```python
df.sort_values(by = ["t"], inplace = True)
df = df.set_index("t")
df
```

**Filling 1st Three values in a column(temperature_of_dry_end) which are missing by using "fillna" Function** 

```python
df['temperature_of_dry_end'].fillna(method='ffill', limit=3, inplace=True)
df
```

**Filling The Missing Values in A whole Dataframe Using "Fillna" Function.**

```python
df.fillna(method='ffill', inplace=True)
df
```

**Importing Train DataFrame**

```python
DataFrame = pd.read_csv("clean_sensor_data_local_time_zone.csv")
DataFrame
```

```python
DataFrame.info()
```

```python
DataFrame.isnull().sum()
```

```python
DataFrame["t"] = pd.to_datetime(DataFrame["t"])
DataFrame.info()
```

```python
DataFrame.sort_values(by = ["t"], inplace = True)
DataFrame = DataFrame.set_index("t")
DataFrame
```

```python
DataFrame["grammage_valve_opening"].fillna(method = "bfill", inplace = True)
DataFrame["pope_speed"].fillna(method = "bfill", inplace = True)
DataFrame["stock_pump_flow"].fillna(method = "bfill", inplace = True)
DataFrame["temperature_of_dry_end"].fillna(method = "bfill", inplace = True)
DataFrame["temperature_of_wet_end"].fillna(method = "bfill", inplace = True)
DataFrame["yankee_speed"].fillna(method = "bfill",  inplace = True)
DataFrame["yankee_steam_pressure"].fillna(method = "bfill", inplace = True)
DataFrame
```

```python
DataFrame.fillna(method = "ffill", inplace = True)
DataFrame
```

**Making Line Plot For checking the Data**

```python
plt.figure(figsize = (18, 7))
sns.lineplot(x = "t", y ="grammage_valve_opening", data = DataFrame, color = "blue")
plt.title("Date/Time across Grammage Value", weight = "bold", size = 18)
plt.xticks(size = 15)
plt.yticks(size = 15)
plt.xlabel("Date/Time", weight = "bold", size = 15)
plt.ylabel("Gammage Values Opening", weight = "bold", size = 15)
plt.show()
```
